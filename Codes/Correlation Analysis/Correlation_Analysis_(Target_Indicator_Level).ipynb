```python
#Read in data
Indonesia=pd.read_csv('https://github.com/PeishanLi/G5055_Practicum_Project2/raw/main/Data/processedIndo.csv')
Guatemala=pd.read_csv('https://github.com/PeishanLi/G5055_Practicum_Project2/raw/main/Data/processedGuate.csv')
Imeasures=pd.read_csv('https://github.com/PeishanLi/G5055_Practicum_Project2/raw/main/Data/measure_picked/Indo_measureIDs.csv')
Gmeasures=pd.read_csv('https://github.com/PeishanLi/G5055_Practicum_Project2/raw/main/Data/measure_picked/Gua_measureIDs.csv')

#A little preprocessing of data
Indonesia=Indonesia.drop('Unnamed: 0', axis=1)
Indonesia['UniqueIDNew']=[x.replace(" ","") for x in Indonesia['UniqueID']]
Guatemala=Guatemala.drop('Unnamed: 0', axis=1)
Guatemala['UniqueIDNew']=[x.replace(" ","") for x in Guatemala['UniqueID']]
Imeasures['xNew']=[x.replace(" ","") for x in Imeasures['x']]
Gmeasures['xNew']=[x.replace(" ","") for x in Gmeasures['x']]
```

## Filter data

Measure the measures picked using Rmd file with the original dataset, to identify which indicator of which target and goal it belongs to, and filters the original dataset for further correlation analysis at target-indicator level.

Method 1

```python
FilteredIndo=pd.DataFrame()
FilteredGuate=pd.DataFrame()
```

```python
def filter_data(df, filtered):
  for i in range(len(df)):
    if df['UniqueIDNew'][i] in Imeasures['xNew']:
      filtered=filtered.append(df.loc[i])
  #filtered.to_csv('/content/drive/MyDrive/Colab Notebooks/Practicum SDG Networks/Data/Indonesia data after selecting one measurement for each indicator.csv')
  #print('Filtering completed!')
  return filtered
```

```python
filter_data(Indonesia, FilteredIndo)
```

## Calculate correlation among indicators

```python
def calculate_correlation(df):
  Alltargets=df['Target'].drop_duplicates()
  for target in Alltargets:
    Allindicators=list(df['Indicator'].drop_duplicates())
    if len(Allindicators)>1:
      group=df[df['Target']==target]
      group=group.drop(columns=['UniqueID','Goal','Target','SeriesCode','UniqueIDNew'],axis=1)
      group_pivot=group.pivot(index=['TimePeriod'], columns=['Indicator'], values='Value').reset_index()
      group_pivot=group_pivot.apply(lambda x:x.astype(float))
      correlation=group_pivot.corr()
      correlation=pd.DataFrame(correlation)
      correlation=correlation.drop('TimePeriod', axis=1)
      correlation=correlation.drop('TimePeriod', axis=0)
      path='/content/drive/MyDrive/Colab Notebooks/Practicum SDG Networks/Data/Indonesia Correlation among Indicators/Indonesia Correlation among Indicators of Target '+target+'.csv'
      correlation.to_csv(path)
      print('Target '+target+' completed')
```

```python
calculate_correlation(FilteredIndo)
```

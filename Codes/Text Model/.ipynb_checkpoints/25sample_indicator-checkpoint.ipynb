```python
#246*5 rows 246indicator, top5 related indicator
all_indicators = pd.read_excel('method_outcome.xlsx')
all_indicators = all_indicators.iloc[:, 0:7]
all_indicators
```

```python
# 17goal 25 indicator, 1 goal 1-2indicators
sample = ['1.1.1', '1.3.1', '2.1.1', '3.1.2', '3.6.1', '4.4.1', '4.7.1', '5.2.1', '5.5.1', '6.6.1', '7.1.2', '8.1.1', '8.4.1', '9.3.1', 
'10.1.1', '11.2.1', '11.3.1', '12.2.1', '12.8.1', '13.3.1', '14.4.1', '15.1.1', '15.4.1', '16.1.2', '17.1.1']
#25 indicators selected
sample = all_indicators[all_indicators['indicator'].isin(sample)]

#sample.to_excel('sample_125rows.xlsx', index = False)
sample
```

```python
sample_noscore = sample[['indicator', 'related_indicator_tf_idf', 'related_indicator_doc2vec', 'related_indicator_word2vec']]
sample_noscore
```

```python
tf_idf = sample_noscore.groupby('indicator')['related_indicator_tf_idf'].apply(',  '.join).to_frame().reset_index()
doc2vec = sample_noscore.groupby('indicator')['related_indicator_doc2vec'].apply(',  '.join).to_frame().reset_index()
word2vec = sample_noscore.groupby('indicator')['related_indicator_word2vec'].apply(',  '.join).to_frame().reset_index()
```

```python
sample_25 = tf_idf.merge(doc2vec, on = 'indicator').merge(word2vec, on = 'indicator')
sample_25
#change the order of indicator in excel sheet
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.stats.stats import pearsonr
sc = StandardScaler()
import itertools
!pip install pingouin
import pingouin as pg
```

```python
#Indonesia=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Practicum SDG Networks/Data/indo_pc1s.csv')
Indonesia=pd.read_csv('https://raw.githubusercontent.com/PeishanLi/G5055_Practicum_Project2/main/Data/PCA_results/indo_pc1s.csv')
Indonesia=Indonesia.rename(columns={'Unnamed: 0': 'Year'})
Indonesia=Indonesia.set_index('Year')
Indonesia_scaled=sc.fit_transform(Indonesia)
Indonesia_scaled=pd.DataFrame(Indonesia_scaled, columns=Indonesia.columns)
Indonesia
```

```python
Indonesia_scaled
```

```python
# Initial Correlations Run: Standard library 

Indonesiacorrelation=Indonesia.corr()
Indonesiascaledcorrelation=Indonesia_scaled.corr()
print(Indonesiacorrelation)
print(Indonesiascaledcorrelation)
Indonesiacorrelation.to_csv('../../Data/PCA_Results/Indonesia_correlation_among_indicators_indicator_model.csv')
Indonesiascaledcorrelation.to_csv('../../Data/PCA_Results/Indonesia_scaled_correlation_among_indicators_indicator_model.csv')

# Also tried to use scipy pearson to look at pearson with p value
# correlations = {}
# columns = Indonesia_scaled.columns.tolist()
# for col_a, col_b in itertools.combinations(columns, 2):
#     correlations[col_a + '__' + col_b] = pearsonr(Indonesia_scaled.loc[:, col_a], Indonesia_scaled.loc[:, col_b])

# result = pd.DataFrame.from_dict(correlations, orient='index')
# result.columns = ['PCC', 'p-value']
# result

# a package called pingouin does both :) just pip install this package if you are running it on your machine (see code above)
Indonesia_corr_p_val = pg.pairwise_corr(Indonesia, method='pearson')
Indonesia_corr_p_val.to_csv('../../Data/PCA_Results/Indonesia_correlation_among_indicators_indicator_model_pval.csv')
print(Indonesia_corr_p_val)
Indonesia_scaled_corr_p_val = pg.pairwise_corr(Indonesia_scaled, method='pearson')
Indonesia_scaled_corr_p_val.to_csv('../../Data/PCA_Results/Indonesia_scaled_correlation_among_indicators_indicator_model_pval.csv')
print(Indonesia_scaled_corr_p_val)

# p-unc is the p value. 
```

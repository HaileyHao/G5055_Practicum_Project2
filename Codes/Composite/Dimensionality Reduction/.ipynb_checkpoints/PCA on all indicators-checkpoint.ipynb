```python
import os
import numpy as np
import pandas as pd

# a function that get all filenames in a folder as a list
def get_all_files(folder):
    file_list = []
    if os.path.exists(folder):
        for root, dirs, files in os.walk(folder):
            for file in files:
                file_list.append(os.path.join(root,file))
    return file_list


country = "Indonesia"
# imputed_prefix = "../../Documents/GitHub/G5055_Practicum_Project2/Data/Data_preprocessed_for_PCA/Indicators_with_imputation/"

imputed_prefix = "../../../Data/Data_preprocessed_for_PCA/Indicators_with_imputation/"

p = f"{imputed_prefix}{country}"
# imputed data filename list
filename = get_all_files(p)
# filename
len(filename)
```

```python
# store pca result: the first PC in a dataframe
def pc1(imputed_d):
    from sklearn.preprocessing import scale
    X = pd.DataFrame(scale(imputed_d), index=imputed_d.index, columns=imputed_d.columns)
#     print(X)
    from sklearn.decomposition import PCA
    # PCA(n_components=None, *, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)
    pca = PCA(n_components = 1).fit(X) # n_components
    X_plot = pd.DataFrame(pca.fit_transform(X), columns=['PC1'], index=X.index)
#     print(pca.explained_variance_ratio_ )
    # return 
    return X_plot["PC1"]
```

```python
# apply pca to all indicators
df = pd.DataFrame(index = list(range(2012,2021)))
import re
for i in filename:
    d = pd.read_csv(i, index_col = "Year").iloc[:-1].drop(["Unnamed: 0"], axis = 1)  # delete 2021
#     print(d)
    # regex to get the indicator index
    prefix = "../../../Data/Data_preprocessed_for_PCA/Indicators_with_imputation/Indonesia/Indicator"
    suffix = ".csv"
    i = re.sub(prefix, '', i)
    i = re.sub(suffix, '', i)
    print(i)
#     print(pc1(d))
    df[i] = pc1(d).values # ndarray
```

```python
def pc1_variance(imputed_d):
    from sklearn.preprocessing import scale
    X = pd.DataFrame(scale(imputed_d), index=imputed_d.index, columns=imputed_d.columns)
#     print(X)
    from sklearn.decomposition import PCA
    # PCA(n_components=None, *, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)
    pca = PCA(n_components = 1).fit(X) # n_components
#     print(pca.explained_variance_ratio_ )
    # return 
    return pca.explained_variance_ratio_ 
```

```python
path = "../../../Data/Data_preprocessed_for_PCA/Indicators_with_imputation/Indonesia/Indicator7.a.1.csv"
d = pd.read_csv(path, index_col = "Year").iloc[:-1] # delete 2021
pc1_variance(d)[0]
```

```python
import re
indicators = []
evr = []
for i in filename:
    d = pd.read_csv(i, index_col = "Year").iloc[:-1].drop(["Unnamed: 0"], axis = 1)  # delete 2021
    # regex to get the indicator index
    prefix = "../../../Data/Data_preprocessed_for_PCA/Indicators_with_imputation/Indonesia/Indicator"
    suffix = ".csv"
    i = re.sub(prefix, '', i)
    i = re.sub(suffix, '', i)
    print(i)
    indicators.append(i)
    print(pc1_variance(d)[0])
    evr.append(pc1_variance(d)[0])
```

```python
df2 = pd.DataFrame()
df2["indicators"] = indicators
df2["evr"] = evr
```

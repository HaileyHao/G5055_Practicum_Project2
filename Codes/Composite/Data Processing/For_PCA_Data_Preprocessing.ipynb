Create data processing function

```python
def PCA_data_processing(countrydf):
  Allindicators=countrydf['Indicator'].drop_duplicates()
  for indicator in Allindicators:
    
    #Create a dataframe for each indicator, where rows are years and columns are different measurements
    df=pd.DataFrame()
    Allmeasurements=countrydf[countrydf['Indicator']==indicator]['SeriesCode'].drop_duplicates()
   
    #Add Year column
    #df['Year']=['2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
    #In Indonesia & Guatemala dataframes, type of values in TimePeriod is float64
    df['Year']=[2012.0,2013.0,2014.0,2015.0,2016.0,2017.0,2018.0,2019.0,2020.0,2021.0]
    year=list(df['Year'])
    
    #Add the columns  
    df=pd.concat([df,pd.DataFrame(columns=Allmeasurements)])
    
    #Set index as Year to prepare it for PCA now that we have the dataframe framework
    df=df.set_index('Year')
    
    #Fill in values by looking back into countrydf
    for i in year:
      for j in df.columns:
        valuelist=countrydf[(countrydf['Indicator']==indicator) & (countrydf['TimePeriod']==i) & (countrydf['SeriesCode']==j)]['Value'].values.tolist()
        #Fill in value only the list is not empty, otherwise the value is NaN
        if len(valuelist)!=0:
          df[j][i]=valuelist[0]
    
    #There are some measurements and indicators that do not have much data, so we filtered the measurements and indicators that will be kept.
    num_rows_without_values=sum([True for idx,row in df.iterrows() if all(row.isnull())])
    if num_rows_without_values<9:
      print('Indicator '+indicator+' have values for more than one year so the dataframe is kept for further analysis!')
      #df.to_csv('/content/drive/MyDrive/Colab Notebooks/Practicum SDG Networks/Data/data processing for PCA/Indonesia/Indicator'+indicator+'.csv')
      #For Guatemala it will be (remember to rerun function code trunk)
      df.to_csv('/content/drive/MyDrive/Colab Notebooks/Practicum SDG Networks/Data/data processing for PCA/Guatemala/Indicator'+indicator+'.csv')
    else: 
      print('Indicator '+indicator+' only have values for one year so the dataframe is not exported!')
    
    #df.to_csv('/content/drive/MyDrive/Colab Notebooks/Practicum SDG Networks/Data/data processing for PCA/Guatemala original/Indicator'+indicator+'.csv')
    #print('Indicator '+indicator+' completed!')
```

Apply function

```python
PCA_data_processing(Indonesia)
```

```python
PCA_data_processing(Guatemala)
```
